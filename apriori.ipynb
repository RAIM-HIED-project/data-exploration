### Algorytm Apriori

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_apriori = df.copy()
df_apriori.head()
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['sex'])
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['blood_group'])
```

```python
df_apriori.drop(['user_id', 'illnesses_input', 'hb', 'hct', 'mch', 'mchc', 'mcv', 'rcc', 'rdw', 'wbc', 'basophil', 'eosinophil', 'lymphocyte', 'monocyte', 'neutrophil', 'platelets'], axis=1, inplace=True)
```

```python
df_apriori.loc[df_apriori['frequency'] == 1.0, 'frequency'] = 'once_a_year'
df_apriori.loc[df_apriori['frequency'] == 2.0, 'frequency'] = 'less_than_once_a_year'
df_apriori.loc[df_apriori['frequency'] == 3.0, 'frequency'] = 'hard_to_say'
df_apriori.loc[df_apriori['frequency'] == 0.0, 'frequency'] = 'more_than_once_a_year'
df_apriori['frequency'] = df_apriori['frequency'].fillna('hard_to_say')
```

```python
df_apriori.loc[df_apriori['covid'] == 1.0, 'covid'] = 'two_weeks_from_covid'
df_apriori.loc[df_apriori['covid'] == 2.0, 'covid'] = 'more_than_two_weeks'
df_apriori.loc[df_apriori['covid'] == 3.0, 'covid'] = 'dont_know'
df_apriori.loc[df_apriori['covid'] == 0.0, 'covid'] = 'no_covid'
df_apriori['covid'] = df_apriori['covid'].fillna('dont_know')
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['frequency'])
df_apriori = pd.get_dummies(data=df_apriori, columns=['covid'])
```

```python
df_test = df_apriori.copy()
df_test['height_cat'] = pd.cut(x=df_test.height, bins=(100, 170, 180, 195), labels=['short', 'medium', 'high'])
df_test
```

```python
df_apriori['age'] = pd.cut(x=df_apriori.age, bins=3, labels=['young', 'mid', 'older'])
df_apriori.head()
```

```python
df_apriori['weight'] = pd.cut(x=df_apriori.weight, bins=3)
df_apriori.head()
```

```python
df_apriori['height'] = pd.cut(x=df_apriori.height, bins=(100, 170, 180, 195), labels=['short', 'medium', 'high'])
df_apriori.head()
```

```python
df_apriori.isnull().sum()
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['age'])
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['weight'])
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['height'])
```

```python
df_apriori['all_parameters_ok'] = df_apriori['all_parameters_ok'].apply(lambda x: 'yes' if x == 1 else 'no')
```

```python
df_apriori['chronic_illness'] = df_apriori['chronic_illness'].apply(lambda x: 'yes' if x == 1 else 'no')
```

```python
df_apriori.head(10)
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['all_parameters_ok'])
```

```python
df_apriori = pd.get_dummies(data=df_apriori, columns=['chronic_illness'])
```

```python
df_apriori.head()
```

```python
df_apriori.drop(['frequency_hard_to_say', 'covid_dont_know'], axis=1, inplace=True)
```

```python
supports = apriori(df_apriori, min_support=0.3, use_colnames=True)
```

```python
rules = association_rules(supports, metric='confidence', min_threshold=0.2)
rules
```
